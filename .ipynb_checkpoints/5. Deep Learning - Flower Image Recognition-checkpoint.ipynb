## <font color = 'blue'> I. Get the data </font>

```python
### IMPORTS ###
import pandas as pd
import numpy as np
import os
# Plotting
%matplotlib inline
import matplotlib.pyplot as plt
# CNN
from tensorflow import keras
from keras.preprocessing import image_dataset_from_directory
from tensorflow.keras import layers
```

```python
# One subfolder by category of flower
train_path = os.path.join("data/flowers")
classes = os.listdir(train_path)
```

```python
# Determine the length of the dataset, as well as the number of flowers per category
count = 0
for clas in classes:
    count = count+len(os.listdir(train_path+"/"+clas))
    print("Number of {} : {}".format(clas, len(os.listdir(train_path+"/"+clas))))
print("Total number of samples: {}".format(count))
print("Flower categories: {}".format(classes))
```

```python
# Preprocess the data:
# This 'image_dataset_from_directory' function allows to directly extract all images from subfolders located in a
# specific folder. It only takes images, so no need to remove all non-jpeg files.
# Images are automatically resized to 256x256.

df = image_dataset_from_directory("data/flowers", labels='inferred', image_size=(256, 256))
```

```python
# Visualising the first 9 images

plt.figure(figsize=(10, 10))
for images, labels in df.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(int(labels[i]))
        plt.axis("off")
```

## <font color = 'blue'> II. Convolution Neural Network model </font>
### <font color = 'blue'>II.a. Create test & train sets </font>

```python
# Parameters
image_size=(256, 256)
batch_size=32
num_classes=5
```

```python
# Generate train and test data sets
train_ds = image_dataset_from_directory(
    "data/flowers", 
    labels='inferred', 
    label_mode='categorical',
    validation_split=0.2,
    subset="training",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
)
val_ds = image_dataset_from_directory(
    "data/flowers",
    labels='inferred',
    label_mode='categorical',
    validation_split=0.2,
    subset="validation",
    seed=1337,
    image_size=image_size,
    batch_size=batch_size,
)
```

### <font color = 'blue'>II.b. Simple model as seen in class </font>

```python
# Define a function that will recreate the model as seen in the course

def make_simple_model(input_shape, num_classes):
    inputs = keras.Input(shape=input_shape)
    x = inputs

    # Entry block: 5x5 convolution with 2x2 stride and 32 filters
    x = layers.experimental.preprocessing.Rescaling(1.0 / 255)(x)
    x = layers.Conv2D(32, (5,5), strides=(2,2), padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    # Another 5x5 convolution with 2x2 stride and 32 filters
    x = layers.Conv2D(32, (5,5), padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)
                             
    ## 2x2 max pooling reduces to 3 x 3 x 32
    x = layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = layers.Flatten()(x)
    x = layers.Dense(512)(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return keras.Model(inputs, outputs)

simple_model = make_simple_model(input_shape=image_size + (3,), num_classes=num_classes)
```

```python
# Print out model summary for ease of reading

simple_model.summary()
```

```python
# Train the model

epochs = 15

callbacks = [
    keras.callbacks.ModelCheckpoint("save_at_{epoch}.h5"),
]
simple_model.compile(
    optimizer=keras.optimizers.RMSprop(lr=0.0005, decay=1e-6),
    loss="categorical_crossentropy",
    metrics=["accuracy"],
)
simple_model.fit(
    train_ds, epochs=epochs, callbacks=callbacks, validation_data=val_ds,
)
```

Training this simple model on 15 epochs takes c. 2 hours.
The achieved validation accuracy is c. 50%

Let's see if we can improve this accuracy by making a more complex model.

### <font color = 'blue'>II.c. More complex model </font>

```python
# Define a function that creates a more complex model

def make_complex_model(input_shape, num_classes):
    inputs = keras.Input(shape=input_shape)
    x = inputs

    # Entry block
    x = layers.experimental.preprocessing.Rescaling(1.0 / 255)(x)
    x = layers.Conv2D(32, 3, strides=2, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.Conv2D(64, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    previous_block_activation = x  # Set aside residual

    for size in [128, 256, 512]:
        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(size, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.Activation("relu")(x)
        x = layers.SeparableConv2D(size, 3, padding="same")(x)
        x = layers.BatchNormalization()(x)

        x = layers.MaxPooling2D(3, strides=2, padding="same")(x)

        # Project residual
        residual = layers.Conv2D(size, 1, strides=2, padding="same")(
            previous_block_activation
        )
        x = layers.add([x, residual])  # Add back residual
        previous_block_activation = x  # Set aside next residual

    x = layers.SeparableConv2D(1024, 3, padding="same")(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("relu")(x)

    x = layers.GlobalAveragePooling2D()(x)
    if num_classes == 2:
        activation = "sigmoid"
        units = 1
    else:
        activation = "softmax"
        units = num_classes

    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(units, activation=activation)(x)
    return keras.Model(inputs, outputs)

complex_model = make_complex_model(input_shape=image_size + (3,), num_classes=num_classes)
```

```python
# Print out model summary for ease of reading

complex_model.summary()
```

```python
# Train the model

epochs = 15

callbacks = [
    keras.callbacks.ModelCheckpoint("save_at_{epoch}.h5"),
]
complex_model.compile(
    optimizer=keras.optimizers.RMSprop(lr=0.0005, decay=1e-6),
    loss="categorical_crossentropy",
    metrics=["accuracy"],
)
complex_model.fit(
    train_ds, epochs=epochs, callbacks=callbacks, validation_data=val_ds,
)
```

Training this more complex model on 15 epochs takes c. 5 hours. The achieved validation accuracy is c. 70%, which is an improvement compared to the 50% of the simple model.

### <font color = 'blue'> III.b. Improvements </font>

We can distinguish 2 types of improvements:
1. Model training time
2. Accuracy


For model training time improvements, we could:
- Import images in greyscale instead of colours, although this could impair accuracy
- Simplify the models by removing layers, or implementing bigger strides (to the potential detriment of accuracy)
- Balance a bit more training and validation sets (e.g. 70/30%)
- Use Google Collab 

For accuracy improvements, we could:
- Train models on more epochs, although I would need a more powerful machine
- Reduce overfitting by using image data augmentation, which artificially introduce sample diversity by applying transformations to the training images, such as random horizontal flipping or small random rotations
- Increase the size of the dataset by scraping more flower images
