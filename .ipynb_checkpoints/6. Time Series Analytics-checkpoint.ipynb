```python
### IMPORTS ###
import pandas as pd
import numpy as np
import os
from datetime import datetime, timedelta
# Plotting
%matplotlib inline
import matplotlib.pyplot as plt
# Stats
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, month_plot
from statsmodels.tsa.seasonal import seasonal_decompose
import pmdarima as pm
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
df = pd.read_csv('data/fake_dataset.csv')
```

```python
# First, let's unstack the dataframe in order to have 1 row per day and 1 column per country.

df_pivot = df.pivot(index='order_date', columns='country', values='sales')
df_pivot.fillna(0, inplace=True)
```

```python
df_pivot.tail()
```

```python
df_pivot.index = pd.to_datetime(df_pivot.index)
```

```python
# Then, let's add missing dates in order to have a continuous time period (i.e. no missing days)

# Make date_order a date instead of a string
df_pivot.index = pd.to_datetime(df_pivot.index)

print('Unique dates in our data: ', len(df_pivot.index.unique()), 'Days')
our_date_range = df_pivot.index.max() - df_pivot.index.min() + timedelta(days=1)

# Calculate number of days in date range
print('Total days in our date range:', our_date_range.days, 'Days')
```

No need to add any missing day in the dataframe we are working with.

##### <font color = 'blue'> 2.b. Exploratory Data Analysis </font>

```python
# Let's prepare our data by separating 2 datasets: a daily one and a weekly one.
df_daily = df_pivot
df_weekly = df_pivot.resample('W').sum()
```

Fakeistan seems to be the primary market for product A.
Now let's have a look at historical trends by top 5 country over the latest year on a:

- Weekly basis

* We can clearly see some weekly seasonality on the daily graph, meaning that the sales are likely to be influenced by the day of the week.

* There is also a sharp daily drop on the Fakeistan daily graph likely to be caused by Christmas lack of sales.

Now let's take a look at a monthly seasonal graph to confirm if there is also a monthly seasonality effect.

```python
#Plot monthly seasonality
df.set_index('order_date', inplace=True)
df.index = pd.to_datetime(df.index)
m_plot = month_plot(df.resample('M').sum()['sales'])
```

It is not clear that there is a monthly seasonality effect, as the very strong trend might influence this plot more than the underlying seasonality (i.e. our dataset stops in May 2017, and given the strong upwards trend, if the growth continued for the rest of 2017 we might have had higher monthly averages for June-December).

##### <font color = 'blue'> 2.c. Time Series Decomposition </font>

Let's decompose the time series in trend, seasonality and residuals.

```python
# Add a column summing all the columns
df_daily['sum'] = df_daily.sum(axis=1)
df_daily = df_daily['2016':]

# Decompose the series into trend, seasonal and residuals
ss_decomposition = seasonal_decompose(x=df_daily['sum'], model='multiplicative')
est_trend = ss_decomposition.trend
est_seasonal = ss_decomposition.seasonal
est_residual = ss_decomposition.resid
```

```python
# Plot the three graphs in addition to the original summed dataframe
fig, axes = plt.subplots(4, 1, sharex=True, sharey=False)
fig.set_figheight(10)
fig.set_figwidth(15)

axes[0].plot(df_daily['sum'], label='Original')
axes[0].legend(loc='upper left');

axes[1].plot(est_trend, label='Trend')
axes[1].legend(loc='upper left');

axes[2].plot(est_seasonal, label='Seasonality')
axes[2].legend(loc='upper left');

axes[3].plot(est_residual, label='Residuals')
axes[3].legend(loc='upper left');
```

Those plots confirm that the trend component is very strong (3x in 17 months), and there is also a seasonality component, probably on a day-of-week basis.

Finally, let's plot autocorrelation and partial autocorrelation.

```python
# Weekly data
df_weekly['sum'] = df_weekly.sum(axis=1)
fig = plot_acf(df_weekly['sum'], lags=range(1,30))
fig = plot_pacf(df_weekly['sum'], lags=range(1,30))
```

```python
# Daily data
fig = plot_acf(df_daily['sum'], lags=range(1,30))
fig = plot_pacf(df_daily['sum'], lags=range(1,30))
```

There is clearly trend and seasonality affecting our data set. Because of those two strong components, let's try and forecast our sales using ARIMA in order to account for them (more basic ARMA would need trend and seasonality removed).

## <font color = 'blue'> 3. Forecasting sales </font>
##### <font color = 'blue'>3.a. ARIMA model </font>

Let's try to fit an ARIMA model by using pmdarima.auto_arima.
* Let's take the weekly data in order to avoid day seasonality, and let the model otpimise d (the order of differencing). We suspect there won't be any seasonality as we didn't see anything in the plot above on the weekly data.
* Let's also try and predict the total sales instead of sales by country in order to simplify the exercise as a first step. We can add a decomposition by country in the improvements part.

```python
df_weekly = df_pivot.resample('W').sum()
df_weekly = df_weekly['2015':]
auto_model = pm.auto_arima(df_weekly['sum'], start_p=0, start_q=0,
                           max_p=10, max_q=10, m=1,
                           start_P=0, seasonal=True,
                           d=None, D=None, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)
```

```python
print('order: ',auto_model.order)
print('seasonal order: ',auto_model.seasonal_order)
```

As expected, there is no seasonality detected with this model. 
* There is a second degree differencing in order to get rid of the trend
* No AR(p), but a MA(q) of order 1.

```python
sar4 = SARIMAX(df_weekly['sum'], 
                    order=(0,2,1), 
                    seasonal_order=(0,0,0,0), 
                    trend='c').fit()
```

```python
sar4.plot_diagnostics(lags=24,figsize = (20,10))
```

The data is not really normal as confirmed by QQplot and histogram, however the autocorrelation is 0.
Let's now visualise predicted vs. actual sales in order to grasp if our model does a good job.

##### <font color = 'blue'>3.b. Validation </font>

```python
def cross_validate(series,horizon,start,step_size,order = (1,0,0),seasonal_order = (0,0,0,0),trend=None):
    '''
    Function to determine in and out of sample testing of arima model    
    
    arguments
    ---------
    series (seris): time series input
    horizon (int): how far in advance forecast is needed
    start (int): starting location in series
    step_size (int): how often to recalculate forecast
    order (tuple): (p,d,q) order of the model
    seasonal_order (tuple): (P,D,Q,s) seasonal order of model
    
    Returns
    -------
    DataFrame: gives fcst and actuals with date of prediction
    '''
    fcst = []
    actual = []
    date = []
    for i in range(start,len(series)-horizon,step_size):
        model = SARIMAX(series[:i+1], #only using data through to and including start 
                                order=order, 
                                seasonal_order=seasonal_order, 
                                trend=trend).fit()
        fcst.append(model.forecast(steps = horizon)[-1]) #forecasting horizon steps into the future
        actual.append(series[i+horizon]) # comparing that to actual value at that point
        date.append(series.index[i+horizon]) # saving date of that value
    return pd.DataFrame({'fcst':fcst,'actual':actual},index=date)
```

```python
#warnings.filterwarnings("ignore")
series = df_weekly['sum']
horizon = 3
start = int(len(df_weekly)*.75)
step_size = 1
order = auto_model.order
seasonal_order = auto_model.seasonal_order

cv1 = cross_validate(series,horizon,start,step_size,
                    order = order,
                    seasonal_order = seasonal_order)
```

```python
cv1.head()
```

```python
plt.plot(cv1.actual, label='actual sales')
plt.plot(cv1.fcst, label='predicted sales')
plt.legend(loc="upper left")
plt.title('Comparison out-of-sample predictions vs. actual sales with ARIMA');
```

We can see clearly that our model does an okay job at picking up the trend.
